```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def funcion(x,n,w):    
    return x**n * np.abs(np.sin(w*x)) 
    
n = [1,2,3,4]
w = 10
x = np.linspace(0, np.pi, 100)

for value in n:
    plt.plot(x, funcion(x, value, w), label='n={}'.format(value))
plt.legend(title='w={}'.format(w))
plt.show()
```
